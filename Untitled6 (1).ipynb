```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.util import random_noise



# Load a sample grayscale image
original_image = color.rgb2gray(data.coffee())

# Apply Singular Value Decomposition (SVD)
U, S, Vt = np.linalg.svd(original_image, full_matrices=False)

# Function to reconstruct image with the first k singular values
def reconstruct_image(U, S, Vt, k):
    # Use only the first k singular values for reconstruction
    Uk = U[:, :k]
    Sk = np.diag(S[:k])
    Vtk = Vt[:k, :]
    # Reconstruct the image
    return Uk @ Sk @ Vtk

# Display original and reconstructed images with different values of k
fig, axes = plt.subplots(1, 4, figsize=(26, 10))
axes[0].imshow(original_image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

# Reconstruct images with varying levels of k
k_values = [10, 40, 100]  # Singular values for different levels of compression
for i, k in enumerate(k_values):
    reconstructed_image = reconstruct_image(U, S, Vt, k)
    axes[i + 1].imshow(reconstructed_image, cmap='gray')
    axes[i + 1].set_title(f"Reconstructed with k={k}")
    axes[i + 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
noisy_image = random_noise(original_image, mode='gaussian', var=0.01)

# Perform Singular Value Decomposition (SVD) on the noisy image
U_noisy, S_noisy, Vt_noisy = np.linalg.svd(noisy_image, full_matrices=False)

fig, axes = plt.subplots(1, 4, figsize=(26, 10))
axes[0].imshow(noisy_image, cmap='gray')
axes[0].set_title("Noisy Image")
axes[0].axis('off')

# Reconstruct images with different levels of k
k_values = [10, 40, 100]  # Different values for compression levels
for i, k in enumerate(k_values):
    reconstructed_noisy_image = reconstruct_image(U_noisy, S_noisy, Vt_noisy, k)
    axes[i + 1].imshow(reconstructed_noisy_image, cmap='gray')
    axes[i + 1].set_title(f"Reconstructed with k={k}")
    axes[i + 1].axis('off')

plt.tight_layout()
plt.show()
```


**OBSERVATION** We can observe that after adding the gausian noise the quality of the gray scale image has been decresed. as compared to non noise gray scale images the svd is given low quality images. While during recunstruction of the image if we inreases the number of singular component the features retains almost of the original in both with noise and without noise gray scale images.

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from google.colab import drive


image_folder = '/content/target_folder'  # Replace with your actual folder path

# Step 2: Load and preprocess the images
images = []
image_shape = (64, 64)  # Resize images to 64x64 for uniformity

for filename in os.listdir(image_folder):
    img_path = os.path.join(image_folder, filename)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Load image in grayscale
    if img is not None:
        img_resized = cv2.resize(img, image_shape)
        images.append(img_resized.flatten())  # Flatten the 2D image to a 1D array

images = np.array(images)

# Step 3: Standardize the data (optional but recommended for PCA)
images_mean = np.mean(images, axis=0)
images_centered = images - images_mean

# Step 4: Apply PCA for dimensionality reduction
n_components = 10 # Number of principal components to retain
pca = PCA(n_components=n_components)
images_pca = pca.fit_transform(images_centered)  # Reduced dimension representation

# Step 5: Visualize the result
# Display some of the images using their reduced dimension representation
fig, axes = plt.subplots(3, 5, figsize=(15, 6))
axes = axes.ravel()
for i in range(15):
    # Project back to original space to visualize the reconstructed image
    reconstructed = pca.inverse_transform(images_pca[i]) + images_mean
    reconstructed_img = reconstructed.reshape(image_shape)
    axes[i].imshow(reconstructed_img, cmap='gray')
   # axes[i].set_title(f"Image {i + 1}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()

# Step 6: Display variance ratio for each principal component
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.grid()
plt.show()
```
